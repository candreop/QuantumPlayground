```python
import numpy as np

from qiskit import ClassicalRegister
from qiskit import QuantumRegister
from qiskit import QuantumCircuit
from qiskit import Aer
from qiskit import transpile

from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_bloch_multivector
from qiskit.visualization import plot_state_qsphere
```

```python
# Quantum and classical registers of the teleportation circuit

# The unknown state |psi> (in possession of Alice) to be teleported
psi = QuantumRegister(name='\psi', size=1) 

# The |phi+> Bell state shared by Alice and Bob
bell_state = QuantumRegister(name='\phi', size=2) 

# Classical measurements performed by Alice and Bob
bits_measured_by_alice = ClassicalRegister (name='a', size=2)
bits_measured_by_bob   = ClassicalRegister (name='b', size=1)
```

```python
# Create a quantum circuit using the above registers
qcircuit = QuantumCircuit(psi,bell_state,bits_measured_by_alice,bits_measured_by_bob) 
```

```python
# Set the (unknown) \psi> state possessed by Alice to anything
# This is the state that will be teleported to Bob
#
#psi_state = [1,0]
psi_state = [0,1]
#psi_state = [1/np.sqrt(2),-1.j/np.sqrt(2)]
qcircuit.initialize(psi_state,0)
qcircuit.barrier()
```

```python
# Initialise qubits 1,2 in the |0> state
qcircuit.initialize([1,0],1)
qcircuit.initialize([1,0],2)
```

```python
# Plot initial state
plot_bloch_multivector(Statevector(qcircuit))
```

```python
# Prepare qubits 1,2 in the Bell state
qcircuit.h(1)
qcircuit.cx(1,2)
qcircuit.barrier()
```

```python
# Plot current entangled state
#plot_bloch_multivector(Statevector(qcircuit))
plot_state_qsphere(Statevector(qcircuit))
```

```python
# Add operations applied by Alice to the unknown state |psi> 
# and to her own half of the Bell state
qcircuit.cx(0,1)
qcircuit.h(0)
qcircuit.barrier()
```

```python
# Let Alice perform her measurements and communicate results to Bob
# The measurement of |psi> is stored at the classical register a[1]
# The measurement of |phi_0> is stored at the classical register a[0]
qcircuit.measure([1,0],[0,1]) 
qcircuit.barrier()
```

```python
# Bob performs operation on his half of the Bell state,
# conditioned upon the classical measurements communicated by Alice
with qcircuit.if_test((bits_measured_by_alice[0],1)):
    qcircuit.x(2)
with qcircuit.if_test((bits_measured_by_alice[1],1)):
    qcircuit.z(2)
    
# qcircuit.cx(1,2) 
# qcircuit.cz(0,2) 
qcircuit.barrier()
```

```python
# Now let Bob measure his part of the Bell state 
# His measurement is stored at the classical register b
qcircuit.measure([2],[2]) 
```

```python
# Draw the completed diagram
qcircuit.draw(output='mpl')
```

```python
# Create a backend simulator
backend_simulator = Aer.get_backend("aer_simulator")

# Transpile the quantum circuit
# This generates low-level QASM instructions used by the backend
qcircuit_compiled = transpile(qcircuit, backend_simulator)

# Execute the circuit on the simulator
# Use the specified number of iterations (shots)
simulation = backend_simulator.run(qcircuit_compiled, shots=4096)

# Get the simulation results
simulation_result = simulation.result()

# Extact the simulated measurement counts
measurement_counts = simulation_result.get_counts(qcircuit_compiled)
```

```python
# Print the measurement counts {x xx}
# The  are the results of the measurement of {\phi1 \phi0 \psi}
# The state \phi1 will always yield a measurement that is identical to the initial state of the unknown state |psi>
print(measurement_counts)
```
